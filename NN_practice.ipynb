```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# to test the iris dataset, the dataset has 4 features
# so the input layers should have 4 layers
# 3 hidden layers and output layer
# output layer has 3 neurons

# inheriting nn.Module

class model(nn.Module):
  def __init__(self,input=4,h1=8,h2=16,h3=8,op=3):
    super().__init__() # invoking the init() in the nn.Module
    self.fc1 = nn.Linear(input,h1)
    self.fc2 = nn.Linear(h1,h2)
    self.fc3 = nn.Linear(h2,h3)
    self.output = nn.Linear(h3,op)

  def forward(self,x):
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = F.relu(self.fc3(x))
    x = self.output(x)

    return x
```

```python
# random seed value for randomization
torch.manual_seed(41)
# calling model class
Model = model()
```

```python
# loading the data and training the model
import pandas as pd
import matplotlib.pyplot as plt
```

```python

url = 'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
iris_df = pd.read_csv(url)
iris_df.tail()
```

```python
# splitting the feature and target
X = iris_df.drop('variety',axis=1)
y = iris_df['variety']
```

```python
# converting x values to numpy array
X = X.values
y = y.values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y_encoded = le.fit_transform(y)
y_encoded
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y_encoded,test_size = 0.2,random_state= 41)
```

```python
# converting X values to floattensors and y values to longtensors
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
```

```python
X_test
```

```python
y_test
```

```python
# set the loss function and the optimizer
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(Model.parameters(),lr = 0.01)
```

```python
#training the model
epochs = 100
losses = []
for i in range(epochs):
  #go forward to get a prediction
  y_pred = Model.forward(X_train)

  #measure the loss
  loss = loss_fn(y_pred,y_train)

  #keeping track of the losses
  losses.append(loss.detach().numpy())

  #print in every 10 epoch
  if i%10==0:
    print(f'Epoch:{i}-> loss:{loss}')

  #backpropagation
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
#graphing it out
plt.plot(range(epochs),losses)
plt.xlabel('epochs')
plt.ylabel('loss')
```

```python
#evaluating the model with testing data
y_eval = Model.forward(X_test)
loss = loss_fn(y_eval,y_test)
```
